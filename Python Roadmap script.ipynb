```python
import requests
import json
import csv

# Mockaroo API endpoint
api_url = "https://api.mockaroo.com/api/generate.csv"

# Mockaroo API key
api_key = "YOUR_API_KEY"  # Replace with your actual API key

# Load formatted data from JSON file
json_file_path = 'formatted_data.json'
with open(json_file_path, 'r') as json_file:
    formatted_data = json.load(json_file)

# Specify the output CSV file path
output_csv_path = 'mockaroo_output.csv'

# Make API requests for each item in formatted_data
for item in formatted_data:
    params = {
        'key': api_key,
        'count': 1000,  # Number of rows
        'fields': [item],
        'format': 'csv'
    }

    response = requests.get(api_url, params=params)

    # Write the CSV data to the output file
    with open(output_csv_path, 'a', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        
        # Write header only if the file is empty
        if csv_file.tell() == 0:
            csv_writer.writerow(response.text.strip().split('\n')[0].split(','))

        # Write the data rows
        csv_writer.writerows(response.text.strip().split('\n')[1:])

print(f"Output saved to {output_csv_path}")
```
